```python
# Lets do some Feature Engineering
# Feature - Column
# Engineering - Creation
# Feature Engineering is the process of creating new and more informative features from the existing features in the dataset.
# Advantages of Feature Engineering:
# 1. Improved Model Performance: By creating new features, we can provide more information to the model, which can lead to better performance.
# 2. Reduced Overfitting: By creating features that capture the underlying patterns in the data, we can reduce the risk of overfitting.
# 3. Better Interpretability: New features can help in understanding the data better and provide insights into the relationships between different features.
# eg: 
# 1. Height and weight can be combined to create a new feature called Body Mass Index (BMI).
# 2. Date of birth and current date can be combined to create a new feature called Age.
# 3. Latitude and longitude can be combined to create a new feature called Location.
# 4. Date and time can be combined to create a new feature called Timestamp.
# 5. Basic salary, Transport allowance, and House Rent Allowance can be combined to create a new feature called Gross Salary.

# Lets think of what features we can create from the existing features in our Titanic dataset
# 1. SibSp, Parch:
# Sibling = 1 brother
# Spouse = 1 wife
# SibSp = 2
# Parent = 1 father + 1 mother
# Child = 1 son + 1 daughter
# Parch = 4
# We can create a new feature called FamilySize by combining SibSp and Parch.
# FamilySize = SibSp + Parch + 1  # +1 for the passenger themselves
# 2. Name:
# We can extract the title from the Name column and create a new feature called Title.
# 3. Fare:
# We can create a new feature called FarePerPerson by dividing the Fare by the FamilySize.
# 4. Age:
# We can create a new feature called AgeGroup by categorizing the Age into different groups.
# 5. GenderClass: If the Age is less than 18, then the GenderClass is 'Child', else it is as per the "Sex" column.
```

```python
# Lets understand the titles: 'Mr', 'Mrs', 'Miss', 'Master', 'Don', 'Rev', 'Dr', 'Mme', 'Ms',
    #    'Major', 'Lady', 'Sir', 'Mlle', 'Col', 'Capt', 'the Countess',
    #    'Jonkheer'

# 1. 'Mr': This is the most common title and is used for adult males.
# 2. 'Mrs': This title is used for married women.
# 3. 'Miss': This title is used for unmarried women.
# 4. 'Master': This title is used for young boys.
# 5. 'Don': This is a Spanish title of respect, often used for noblemen.
# 6. 'Rev': This title is used for members of the clergy.
# 7. 'Dr': This title is used for individuals with a doctorate degree, regardless of gender.
# 8. 'Mme': This is the French equivalent of 'Mrs' and is used for married women.
# 9. 'Ms': This title is used for women, regardless of marital status.
# 10. 'Major': This title is used for military officers of the rank of Major.
# 11. 'Lady': This title is used for women of noble birth or rank.
# 12. 'Sir': This title is used for men who have been knighted.
# 13. 'Mlle': This is the French equivalent of 'Miss' and is used for unmarried women.
# 14. 'Col': This title is used for military officers of the rank of Colonel.
# 15. 'Capt': This title is used for military officers of the rank of Captain.
# 16. 'the Countess': This title is used for women of noble rank, specifically the wife of a Count.
# 17. 'Jonkheer': This is a Dutch title of nobility, used for men of noble birth.
```

```python
# value_counts on the Title column
df['Title'].value_counts()
```

```python
# Uncommon titles are:
# Dr                7
# Rev               6
# Mlle              2
# Major             2
# Col               2
# the Countess      1
# Capt              1
# Ms                1
# Sir               1
# Lady              1
# Mme               1
# Don               1
# Jonkheer          1

# Lets merge all uncommon titles into Mr, Miss, Mrs, Master
# Logic:
# 1. If the title is 'Dr' or 'Rev' or 'Major' or 'Col' or 'Capt' or 'Don' or 'Sir' or 'Jonkheer', then we will replace it with 'Mr' (assuming they are male)
# 2. If the title is 'Mlle' or 'Ms', then we will replace it with 'Miss'
# 3. If the title is 'Mme' or 'the Countess' or 'Lady', then we will replace it with 'Mrs'
def merge_titles(title):
    if title in ['Dr', 'Rev', 'Major', 'Col', 'Capt', 'Don', 'Sir', 'Jonkheer']:
        return 'Mr'
    elif title in ['Mlle', 'Ms']:
        return 'Miss'
    elif title in ['Mme', 'the Countess', 'Lady']:
        return 'Mrs'
    else:
        return title

df['Title'] = df['Title'].apply(merge_titles)
df['Title'].value_counts()  # Display the value counts of the Title column after merging
```

```python
# Lets bin the Age column into different groups
# Logic:
# 1. If Age < 12, then AgeGroup = 'Child'
# 2. If 12 <= Age < 20, then AgeGroup = 'Teenager'
# 3. If 20 <= Age < 35, then AgeGroup = 'Young Adult'
# 4. If 35 <= Age < 60, then AgeGroup = 'Adult'
# 5. If Age >= 60, then AgeGroup = 'Senior'
# This process is called Binning or Discretization.
def age_group(age):
    if age < 12:
        return 'Child'
    elif 12 <= age < 20:
        return 'Teenager'
    elif 20 <= age < 35:
        return 'Young Adult'
    elif 35 <= age < 60:
        return 'Adult'
    else:
        return 'Senior'
df['AgeGroup'] = df['Age'].apply(age_group)
df['AgeGroup'].value_counts()  # Display the value counts of the AgeGroup column
```
